```python
# multiple conditions using | Or 
import pandas as pd
df = pd.read_csv('datas4.csv',index_col = 'Sn') 
df[(df['Age']>35) | (df['Addres'] == 'Kathmandu')] 
```

```python
# multiple conditions using & And
import pandas as pd
df = pd.read_csv('datas4.csv',index_col = 'Sn') 
df[(df['Age']>35) & (df['Addres'] == 'Kathmandu')] 
```

```python
# to_csv
import pandas as pd
df = pd.read_csv('datas4.csv',index_col = 'Sn') 
dff = df[(df['Age']>35) | (df['Addres'] == 'Kathmandu')] 
dff.to_csv('my_data.csv')
```

```python
# to_json
dff.to_json('my_data.json')
```
